```python
import pandas as pd
import os
```

```python
df_ads_links = pd.read_csv("./Dados_tratados/TabelaAdsLinks_clear.csv")
df_pesquisa_utm = pd.read_csv("./Dados_tratados/TabelaPesquisaUtm_clear.csv")
```

```python
df_merge = pd.merge(df_pesquisa_utm, df_ads_links, on='anuncio', how='left')
```

```python
df_merge.drop(columns=['adslinks_id'], inplace=True)
```

```python
df_merge
```

```python
df_merge['email'] = df_merge['email'].astype('string')
df_merge['canal_origem'] = df_merge['canal_origem'].astype('string')
df_merge['anuncio'] = df_merge['anuncio'].astype('string')
df_merge['publico'] = df_merge['publico'].astype('string')
df_merge['link_anuncio'] = df_merge['link_anuncio'].astype('string')
```

```python
path_folder = './Dados_tratados'

if not os.path.exists(path_folder):
    os.makedirs(path_folder)

path_file = os.path.join(path_folder, 'adsLinks_pesquisaUtm_merge.csv')

df_merge.to_csv(path_file, index=False)
print(f"Arquivo salvo em: {path_file}")
```
